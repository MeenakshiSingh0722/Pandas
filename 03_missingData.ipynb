Missing Data

```python
import numpy as np
import pandas as pd
```

Finding Missing data

```python
data={
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df=pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum() ##colm wise sum krta null values ka
```

```python
df.isna().any() ##btata hai ki null vale hai ya nhi cols mai
```

Removing missing data

```python
df
```

```python
df.dropna()
```

```python
df.dropna(thresh=3) ##atleast 3 non null value
```

Filling the missing data

```python
df.fillna(0)## view point of
```

```python
values={'A':0,'B':100,'C':300,'D':400}
df.fillna(value=values) ##colms wise fill
```

```python
df.fillna(df.mean())
```
