Pandas Series

```python
import numpy as np
import pandas as pd
```

```python
labels=['a','b','c']
my_list=[10,20,30]
arr=np.array([10,20,30])
my_dict={1:10,2:20,3:30}

pd.Series(my_list)
```

```python
pd.Series(my_list,index=labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(my_dict)
```
